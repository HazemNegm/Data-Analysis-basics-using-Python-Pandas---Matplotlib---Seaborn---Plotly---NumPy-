#For Loop

```python
emails = set()
while True:
    try:
        count = int(input("How many email addresses do you want to enter? "))
        break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

for i in range(1, count + 1):
    email = input(f"Enter email {i}: ").strip()
    emails.add(email)
print("\n--- Extracted Usernames ---")
for email in sorted(emails):
    username = email.split("@")[0]
    print(username)
```

```python
students = []
count = int(input("How many students do you want to enter? "))
for i in range(1, count + 1):
    name = input(f"Enter student {i} name: ").strip()
    students.append(name)
print("\n--- Length of Each Name ---")
for name in students:
    print(f"Length of {name}: {len(name)}")
```

#Function

```python
def sum_of_squares_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num ** 2
    return total
print("Enter numbers one by one. Type 'done' when finished.\n")

numbers = []
while True:
    entry = input("Enter a number: ").strip()
    if entry.lower() == "done":
        break
    numbers.append(int(entry))
result = sum_of_squares_even(numbers)
print(f"\nYour numbers   : {numbers}")
print(f"Even numbers   : {[n for n in numbers if n % 2 == 0]}")
print(f"Sum of squares of even numbers: {result}")
```

```python
def my_abs(num):
    if num < 0:
        return -num
    else:
        return num
num = float(input("Enter a number: "))
result = my_abs(num)
if num == int(num):
    print(f"Absolute value of {int(num)}: {int(result)}")
else:
    print(f"Absolute value of {num}: {result}")
```

```python
def celsius_to_fahrenheit(celsius_list):
    fahrenheit_list = []
    for c in celsius_list:
        f = (c * 9/5) + 32
        fahrenheit_list.append(round(f, 1))
    return fahrenheit_list
print("Enter temperatures in Celsius one by one. Type 'done' when finished.\n")

temperatures = []
while True:
    entry = input("Enter a temperature (°C): ").strip()
    if entry.lower() == "done":
        break
    temperatures.append(float(entry))
result = celsius_to_fahrenheit(temperatures)

print(f"\nCelsius temperatures    : {temperatures}")
print(f"Fahrenheit temperatures : {result}")
```
