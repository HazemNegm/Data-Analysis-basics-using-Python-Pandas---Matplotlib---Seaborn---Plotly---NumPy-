```python
import numpy as np

np.random.seed(42)
arr = np.random.randint(10, 101, size=20)
print("Full Array:  ", arr)
print("\nFirst 5 elements:    ", arr[:5])
print("Last 3 elements:     ", arr[-3:])
print("Even index elements: ", arr[::2])
```

```python
import numpy as np
np.random.seed(42)
arr1 = np.random.randint(10, 101, size=20)
np.random.seed(10)
arr2 = np.random.randint(1, 11, size=20)
print("Array 1:", arr1)
print("Array 2:", arr2)
print("\nAddition:       ", arr1 + arr2)
print("Subtraction:    ", arr1 - arr2)
print("Multiplication: ", arr1 * arr2)
print("Division:       ", np.round(arr1 / arr2, 2))
```
