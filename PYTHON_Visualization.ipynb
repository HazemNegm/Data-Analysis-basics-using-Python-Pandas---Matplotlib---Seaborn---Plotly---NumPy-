```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

#Line Chart

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1700, 1600, 1800, 2000]

plt.plot(months, sales , color='red' ,  linewidth=2 , marker='*')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1700, 1600, 1800, 2000]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=months,
    y=sales,
    mode='lines+markers',
    line=dict(color='green', width=2),
    marker=dict(symbol='circle', size=12, color='red')
))

fig.update_layout(
    title="Monthly Sales Trend",
    xaxis_title="Months",
    yaxis_title="Sales"
)

fig.show()
```

#Bar Chart

```python
products = ["Laptops", "Mobiles", "Tablets", "Accessories"]
revenue = [35000, 50000, 15000, 10000]

combined = sorted(zip(revenue, products), reverse=True)
sorted_revenue = [rev for rev, prod in combined]
sorted_products = [prod for rev, prod in combined]

plt.bar(sorted_products, sorted_revenue)
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.show()
```

```python
products = ["Laptops", "Mobiles", "Tablets", "Accessories"]
revenue = [35000, 50000, 15000, 10000]

combined = sorted(zip(revenue, products), reverse=True)
sorted_revenue = [rev for rev, prod in combined]
sorted_products = [prod for rev, prod in combined]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=sorted_products,
    y=sorted_revenue
))

fig.update_layout(
    title="Revenue by Product",
    xaxis_title="Products",
    yaxis_title="Revenue"
)

fig.show()
```

# Pie Chart

```python
languages = ["Python", "Java", "C++", "JavaScript"]
popularity = [40, 25, 20, 15]

plt.pie(popularity , labels=languages, autopct='%1.1f%%', startangle=90)
plt.title('Programming Language Market Share')
plt.show()
```

```python
languages = ["Python", "Java", "C++", "JavaScript"]
popularity = [40, 25, 20, 15]

fig = go.Figure()

fig.add_trace(go.Pie(
    labels=languages,
    values=popularity,
    textinfo='label+percent',
    rotation=90
))
fig.update_layout(title='Programming Language Market Share')
fig.show()
```
